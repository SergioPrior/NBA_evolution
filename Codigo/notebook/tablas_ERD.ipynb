```python
import pandas as pd
```

# Aqui vamos a crear los dataframes para subirlos al ERD de la manera adecuada

```python
nba = pd.read_csv(r"../csv/Nba_Clean.csv")
nba.drop(columns="Unnamed: 0", inplace = True)
```

```python

nba.insert(0, 'id', nba.index+1)
nba.head()

```

```python
columnas_player = ["id","Name", "age", "Team", "Height", "Weight", "Country"]

Player = nba[columnas_player].copy()
Player.rename(columns=({"id":"idPlayer"}), inplace=True)
```

```python
Player.head()
```

```python
# Creamos la Many to Many de equipos y jugadores por su ID.

equipos = ['Houston Rockets', 'Washington Wizards', 'Memphis Grizzlies',
           'L.A.Lakers', 'Denver Nuggets', 'Orlando Magic',
           'Charlotte Hornets', 'Milwake Bucks', 'Detroit Pistons',
           'Portland Trail Blazers', 'Dallas Mavericks', 'Utah Jazz',
           'Oklahoma City Thunder', 'Boston Celtic', 'Indiana Pacers',
           'San Antonio Spurs', 'Miami Heat', 'Atlanta Hawks',
           'Brooklyn Nets', 'L.A.Clippers', 'Golden State',
           'Philadelphia 76ers', 'New York Knicks', 'Toronto Raptors',
           'Phoneix Suns', 'Minnesota Timberwolves', 'Chicago Bulls',
           'Sacramento Kings', 'Cleveland Cavaliers', 'New Orleands Pelicans']

data_equipos = {'id': range(1, len(equipos)+1), 'Team': equipos}

df_equipos = pd.DataFrame(data_equipos)

df_player_teams = pd.merge(Player, df_equipos, left_on='Team', right_on='Team', how='inner')


df_player_teams.drop(columns=["Name", "age","Team", "Height", "Weight", "Country"], inplace=True)

df_player_teams.rename(columns={"id_x":"idPlayer", "id_y":"idClub"}, inplace=True)
```

```python
#Creamos el many to one de Stats

columnas_stats = ["id", "G_Played", "Pts", "Reb", "Asist"]

Stats = nba[columnas_stats].copy()

Player['idStats'] = Stats['id']

Stats

```

```python
columnas_season = ["id","Season","Decade"]

season = nba[columnas_season].copy()

Player['idSeason'] = season['id']

Player
```

```python
df_equipos.to_csv('clubs.csv', index=False)
Stats.to_csv('Stats.csv', index=False)
df_player_teams.to_csv('player_teams.csv', index=False)
season.to_csv('season.csv', index=False)
Player.to_csv('player.csv', index=False)
```
