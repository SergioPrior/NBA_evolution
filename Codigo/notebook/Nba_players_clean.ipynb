```python
import pandas as pd
import numpy as np
```

```python
nba = pd.read_csv(r"../csv/all_seasons.csv")
pd.options.display.max_columns = None
pd.set_option('display.max_rows',50)
nba.head()
```

```python
# limpieza de columnas no necesarias para el estudio.
nba = nba.drop(columns=["Unnamed: 0", "draft_year","draft_round", "draft_number", "net_rating", "college"])
```

```python
#Pasamos a Int los datos que nos hacen falta
nba.player_height = nba.player_height.astype(int)
nba.age = nba.age.astype(int)
nba.player_weight = nba.player_weight.round(1)
nba.usg_pct = nba.usg_pct.round(2)
nba.ts_pct = nba.ts_pct.round(2)
nba.ast_pct = nba.ast_pct.round(2)
nba.oreb_pct = nba.oreb_pct.round(2)
nba.dreb_pct = nba.dreb_pct.round(2)
```

```python
nba.team_abbreviation.unique()
```

```python
# juntamos las franquicias antiguas por las modernas para tener unos datos unicos de los equipos reales.
nba.team_abbreviation=np.where(np.isin(nba.team_abbreviation, ['SEA']), 'OKC', nba.team_abbreviation)
nba.team_abbreviation=np.where(np.isin(nba.team_abbreviation, ['CHA']), 'CHH', nba.team_abbreviation)
nba.team_abbreviation=np.where(np.isin(nba.team_abbreviation, ['NJN']), 'BKN', nba.team_abbreviation)
nba.team_abbreviation=np.where(np.isin(nba.team_abbreviation, ['VAN']), 'MEM', nba.team_abbreviation)
nba.team_abbreviation=np.where(np.isin(nba.team_abbreviation, ['NOK']), 'NOP', nba.team_abbreviation)
nba.team_abbreviation=np.where(np.isin(nba.team_abbreviation, ['NOH']), 'NOP', nba.team_abbreviation)
```
